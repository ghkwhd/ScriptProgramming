### 2. 경로에 해당하는 문자열 1개를 입력 받아 그 안에 디렉토리 경로명과 파일명을 분리하여 리스트로 반환하는 함수 div_path(s)를 작성하시오.
* 인자로 전달하는 문자열은 경로만 들어간다고 가정한다.
* 각 디렉토리와 파일을 구분하는 문자는 '/'로 가정한다.
* 반환하는 리스트의 첫번째 원소는 디렉토리이고 두번째 원소는 파일명이다.
* 다음과 같은 실행 및 출력 결과가 도출되어야 한다.
    * div_path('/usr/local/bin/python')
        * ['/usr/local/bin', 'python']
    * div_path('/home/chulsoo/test.txt')
        * ['/home/chulsoo', 'test.txt']
* [참고] 리스트(l) 내에 새로운 정수값 (예를 들어 10)을 넣는 방법은 l.append(10) 이다.

```python
def div_path(str) :
    for i in range(len(str)-1,0,-1) :
            if(str[i] == '/'):
                break
    list= []
    list.append(str[:i])
    list.append(str[i+1:])
    return list
            
print(div_path('C/Users/name/anaconda3'))
print(div_path('/usr/local/bin/python'))
print(div_path('/home/chulsoo/test.txt'))
```

* 인자로 전달된 문자열의 길이를 이용하여 반복문을 작성한다.
* 반복문에서 range를 작성할 때, 인덱스 고려를 위해 len(str)-1을 하였고 역순으로 돌아가도록 작성하였다.
* 제일 뒤 쪽이 파일명이기 때문에 역순으로 돌리며 마지막에 있는 /의 위치를 찾는다
* 그 후 리스트를 선언하여 처음부터 마지막 / 가 있는 위치까지 슬라이싱과 append를 이용하여 리스트에 넣는다.
* 파일명은 /를 포함하지 않고 넣기 위해 i+1을 해주었고, 문자열을 끝까지 잘라 리스트에 넣고 리스트를 반환한다.
* 반환값이 있어서 함수호출을 하면서 print를 해주어 바로 출력값을 확인할 수 있도록 하였다.

### 3. 두 개의 리스트를 인자로 받아서 그 두 개의 리스트에 대한 '합집합'을 반환하는 함수 list_union(lista, listb)를 작성하시오.
* **집합 자료형 set은 사용하지 않는다.**
* 인자로 전달하는 리스트 2 개에는 정수값만 들어간다고 가정하자.
* 함수 내에서 새로운 리스트를 만들어 그 리스트 내에 인자로 받은 두 리스트의 모든 원소를 넣어 반환한다.
* 반환하는 리스트에는 절대로 중복된 원소가 들어 있으면 안된다 (집합의 조건).
* 반환하는 리스트는 정렬이 되어 있어야 한다.
* 다음과 같은 실행 및 출력 결과가 도출되어야 한다.
    * list_union([1, 2, 3], [1, 2, 4])
        * [1, 2, 3, 4]
    * list_union([-10, -5, 0, -1], [100, 9, 0, 9])
        * [-10, -5, -1, 0, 9, 100]
    * list_union([0, 1, 2], [0, 1, 2])
        * [0, 1, 2]
* [참고] 리스트(l) 내에 새로운 정수값 (예를 들어 10)을 넣는 방법은 l.append(10) 이다.
* [참고] 임의의 정수값 (x)이 리스트 (l) 내에 존재하는지 판단하는 방법은 x in l 이다.

```python
def list_union(lista,listb) :
    list = []
    k = 0
    
    for i in range(len(lista)) :
        list.append(lista[i])
    
    for i in range(len(listb)) :
        list.append(listb[i])
    list.sort()
    
    while(True) :
            if list[k] == list[k+1] :
                list.remove(list[k])
            else :
                k += 1
  
            if k == len(list)-1 :
                break
                
    return list

print(list_union([1, 1, 1, -1, 3, -4, 2, 4, 5, 3], [7, 8, 1, 2, 10, 4]))
```

* 함수 내에서 새로운 리스트를 선언하고 뒤에 while문에서 사용할 변수 하나를 선언한다.
* for 문을 사용하여 인자로 들어온 리스트 두 개의 모든 원소들을 새로운 리스트에 넣는다.
* sort()를 이용하여 리스트를 정렬한다.
* 정렬 후 중복 삭제를 위해 while문을 사용한다.
* 만약 k번째 원소와 k+1 번째 원소가 같다면 k 번째 원소를 삭제하고 k를 증가시키지 않는다.
* k를 증가시키지 않는 이유는 만약 같은 원소가 2개 이상인 경우에 계속 같은자리부터 확인을 하지 않으면 제대로 삭제되지 않기에 k를 증가시키지 않는다. 
* 만약 같지 않다면 k의 값을 1 증가시킨다.
* 증가한 k의 값이 len(list)-1이라면 즉, 마지막 원소를 나타낸다면 반복문을 중단한다.
* 이 역시 반환값이 있어서 함수호출을 하면서 print를 해주어 바로 출력값을 확인할 수 있도록 하였다.

### 4. 두 개의 리스트를 인자로 받아서 그 두 개의 리스트에 대한 '교집합'을 반환하는 함수 list_intersection(lista, listb)와 '차집합'을 반환하는 함수 list_difference(lista, listb)를 작성하시오.
* 모든 가정과 조건은 3번 문제와 동일하다.

```python
def list_intersection(lista, listb):
    list = []

    for i in range(len(lista)):
        list.append(lista[i])

    for i in range(len(listb)):
        list.append(listb[i])
    list.sort()

    while(True) :
        for i in list:
            if list.count(i) == 1:
                list.remove(i)
                break
            else :
                k = i

        if k == list[len(list)-1] :
            break

    k = 0
    while (True):
        if list[k] == list[k + 1]:
            list.remove(list[k])
        else:
            k += 1

        if k == len(list) - 1:
            break

    return list


def list_difference(lista, listb):
    list = []
    
    for i in range(len(lista)) :
        list.append(lista[i])
    
    for i in range(len(listb)) :
        list.append(listb[i])
    list.sort()
    
    k = 0
    while(True) :
            if list[k] == list[k+1] :
                list.remove(list[k])
            else :
                k += 1
  
            if k == len(list)-1:
                break
    
    k = 0
    while(True) :
        for i in range(len(listb)) :
            if list[k] == listb[i] :
                list.remove(list[k])
                break
                
            if listb[i] == listb[len(listb)-1] :
                k += 1
        
        if k == len(list) :
            break

    return list
    
print('[1, 2, 3, 5, 6, 7] 와 [1, 2, 4, 7, 8] 의 교집합 = ',list_intersection([1, 2, 3, 5, 6, 7], [1, 2, 4, 7, 8]))
print('[1, 2, 3, 5, 6, 7] - [1, 2, 4, 7, 8] = ', list_difference([1, 2, 3, 5, 6, 7], [1, 2, 4, 7, 8]))
print('[1, 2, 4, 7, 8] - [1, 2, 3, 5, 6, 7] = ', list_difference([1, 2, 4, 7, 8], [1, 2, 3, 5, 6, 7]))
```

* 두 집합의 교집합을 넣어 반환하는 list를 선언한다.
* 반복문을 이용하여 인자로 전달된 두 개의 리스트를 list에 삽입 후 정렬한다.
* while 반복문을 이용해 항목이 두 개이상이 아닌 항목들을 삭제한다.
* i를 list의 항목으로 반복되게 하고 count를 이용해 항목의 수가 1이면 리스트에서 제거하고 for 반복문을 중단하도록 하였다.
* else의 경우도 설정하여 k의 값과 i의 값이 동일하게 한 뒤 for 반복문이 종료된 뒤 k가 리스트의 마지막 요소와 같다면 반복문이 중단되도록 하였는데 삭제를 진행하며 마지막요소까지 확인했을 때 while 반복문이 종료되도록 설정한 것이다.
* 그 후 중복된 항목들이 리스트에 있기 때문에 하나씩만 남기기 위해 이 역시 while문을 사용하였다.
* 항목이 그 다음 항목과 같아면 인덱스가 더 작은 항목을 삭제한다. 같지 않다면 인덱스를 나타내는 k를 증가시켜 다음 인덱스를 확인한다.
* k가 계속 증가해서 마지막 인덱스까지 도달하게 되면 반복문을 중단하는데 k가 마지막 이전의 인덱스인 경우 마지막 인덱스까지 확인을 마친 상태로 k가 증가하기 때문에 k가 마지막요소에 도달하면 범위 오류가 나기 때문에 len(list) -1로 조건을 설정하였다.  


* 차집합의 경우 A-B 일때 전체집합에서 B 집합을 빼는 형식으로 진행하였다.  
* 위에서 합집합을 구한 것과 마찬가지로 두 개의 리스트를 하나의 리스트에 종합하고 반복문을 이용하여 중복된 항목들을 삭제하였다.
* 그 후 합집합보다 listb의 크기가 더 작기 때문에 while문 내에서 i를 listb의 크기만큼 만복시키면서 list의 항목과 listb의 항목이 같다면 list에서 삭제하도록 조건문을 설정하였고 삭제된다면  for문을 중단하도록 하였다. for문에서 i의 값을 확인했을 때 마지막 요소까지 계속 반복되었다면 같은 요소가 없다는 의미이기 때문에 list의 다음 인덱스를 확인하도록 k의 값을 증가시켰다.
* k의 값이 계속 증가하여 마지막 인덱스를 넘게 된다면 while문을 종료하도록 설정하고 list를 반환하였다.

### 5. 두 개의 양의 정수를 인자로 받아서 해당 범위안에 있는 소수(prime number)의 리스트를 출력하는 함수 print_primenumber(a, b)를 작성하시오.
* 이미 구현된 모듈이나 라이브러리는 사용하지 않는다.
* 함수의 반환(return)은 없으며 함수 내의 끝에서 .format()을 사용하여 다음과 같이 출력한다.
    * print_primenumber(2, 10)
        * 2 이상 10 이하 정수 중 소수 리스트: [2, 3, 5, 7]

```python
def print_primenumber(a,b) :
    list = []
    k = 0
    for i in range(a,b+1) :
        for j in range(2,i+1) :
            if i % j == 0 :
                k = j
                break
        if i == k :
            list.append(i)
    print('{} 이상 {} 이하 정수 중 소수 리스트 : '.format(a,b),list)
            
print_primenumber(11,100)
```

* 함수 내에 빈 리스트를 선언해 후에 소수들을 저장한다.
* 나누어떨어지는 수가 자기 자신인지 확인하기 위해 변수 k를 선언한다.
* 입력받은 a부터 b까지 반복시키고 그 내부에서 나누어 떨어지는 수를 찾기위해 2부터 자기 자신까지 반복하도록 한다.
* 만약 나누어떨어진다면 k에 나누는 수를 대입하고 반복문을 종료시킨다.
* 만약 k가 자기 자신이라면 자기 자신 외에 나누어떨어지는 수가 없다는 의미이기 때문에 append를 이용해 list에 추가한다.
* format(a,b)를 활용하여 숫자의 범위를 출력하고 그 뒤에 리스트를 출력한다.

* 2개 이상의 URL에 대하여 순수한 단어의 개수를 출력하는 예시를 제시할 것

```python
import requests

req = requests.get('https://comic.naver.com/webtoon/weekday')
source = req.text

req2 = requests.get('https://www.weather.go.kr/w/index.do')
source2 = req2.text

# 전달받은 문자열 사이의 구간 삭제
def delete(string, first, second):
    while (True):
        start = string.find(first)  # first에 들어온 단어의 위치 확인
        end = string.find(second)  # end로 들어온 단어의 위치 확인
        if start != -1 and end != -1:
            string = string[:start] + string[end + len(second):]  # end는 시작되는 위치가 나오기 때문에 end는 글자수 만큼 더해준다
        elif start == -1 and end == -1:
            break
    return string

def find_word(str):
    list = []  # 태그를 제거한 리스트
    word = []  # 공백을 기준으로 나눈 리스트
    result = []  # 특수문자, 숫자, 기타 문법을 제거한 최종 리스트
    temp = []  # 단어 저장을 위한 임시 리스트

    # 태그 사이에 문법이 들어가는 경우의 삭제
    # 간혹 대문자로 쓰인 페이지가 있어서 대문자로 이루어진 태그도 구간 삭제
    str = delete(str, '<style', '</style>')  # <style ~ </style> 구간 삭제
    str = delete(str, '<STYLE', '</STYLE>')  # <STYLE ~ </STYLE> 구간 삭제
    str = delete(str, '<script', '</script>')  # <script ~ </script> 구간 삭제
    str = delete(str, '<SCRIPT', '</SCRIPT>')  # <SCRIPT ~ </SCRIPT> 구간 삭제
    str = delete(str, '<!--', '-->')  # 주석처리 구간 삭제

    # HTML 태그 제거
    while (True):
        start = str.find('<')  # <가 있는 위치 확인
        end = str.find('>')  # >가 있는 위치 확인

        if start == -1 and end == -1:  # 만약 < 와 > 가 없다면
            break  # 모든 문자열을 확인한 것이니 반복문 중단

        # <> 내부에 <>가 있는지 확인
        inner = str[start + 1: end+1]     # <>사이를 자른 문자열 생성
        in_start = inner.find('<')  # <가 있는 위치 확인

        if in_start != -1 :      # <> 사이에 <가 있다면 < <> > 형태이므로
            str = str[:in_start] + str[end + 1:]    # 내부의 <> 삭제
            continue

        list.append(str[:start])  # 반복문이 중단되지 않았다면 처음부터 start 전까지의 문자열을 list에 저장
        str = str[end + 1:]  # > 이후의 문자열만 남긴다

    # &gt, &lt 등 삭제
    i = 0
    while (True):
        start = list[i].find('&')  # 시작이 &
        end = list[i].find(';')  # 끝이 ;

        if start != -1 and end != -1:  # 둘 다 발견되었을 경우
            if start < end :    # &가 앞에서, ;가 뒤에서 발견된 경우
                list[i] = list[i][:start] + list[i][end + 1:]  # &와 ; 사이의 구간을 삭제
            else :      # ; 하나만 쓰여서 &가 뒤에서 발견되는 경우 (나중에 특수문자에서 제거된다)
                i += 1  # 다음 항목으로 넘어간다

        else:  # 발견되지 않았을 경우
            i += 1  # 다음 단어 확인

        if i == len(list):  # 마지막 단어까지 확인하면 중단
            break

    # 공백을 통해 임시로 단어 나누기
    for i in range(len(list)):
        temp = list[i].split()  # 리스트의 항목을 반복시켜 공백으로 나누고 임시 리스트에 저장
        for j in range(len(temp)):  # 임시 리스트의 항목 수 만큼 반복
            word.append(temp[j])  # 임시 리스트의 내용들을 word 리스트에 저장

    # 특수문자 제거
    for i in range(len(word)):
        k = 0
        while (True):
            if not word[i][k].isalnum():  # 문자(영어, 한글)나 숫자로만 이루어진 경우가 아닐 때
                if word[i][k] != '' and word[i][k] != ' ':  # 아무것도 없는 항목이거나 공백만 들어있는 것이 아닐 때
                    word[i] = word[i].replace(word[i][k], ' ')  # 해당 위치를 공백으로 변경
                else:  # 아무것도 들어있지 않거나 공백이라면
                    k += 1  # 단어의 다음 위치 확인
            else:  # 문자와 숫자로만 이루어진 경우
                k += 1  # 문자의 다음 위치 확인

            if k == len(word[i]):  # 문자열의 마지막 부분까지 확인했다면
                break  # 반복 중단

    # 공백을 통해 최종적으로 단어 나누기
    for i in range(len(word)):
        temp = word[i].split()  # word 항목을 반복시켜 공백으로 나누고 임시 리스트에 저장
        for j in range(len(temp)):  # 임시 리스트의 항목 수 만큼 반복
            result.append(temp[j])  # 임시 리스트의 내용들을 result 리스트에 저장

    # 숫자만으로 이루어진 항목 제거
    i = 0
    while (True):
        if result[i].isdigit():  # 항목이 숫자로만 구성되어 있다면 True를 반환해서 조건문 실행
            result.remove(result[i])  # 숫자만으로 구성된 항목은 삭제
        else:
            i += 1

        if i == len(result):
            break

    return result

first = find_word(source)
print(first)
print('단어의 개수 : ', len(first))
print()
second = find_word(source2)
print(second)
print('단어의 개수 : ', len(second))
```

* requsets 패키지를 import하고 get()을 통해 URL을 요청하고 이를 가져와 문자열로 사용하였다.
* 두 개의 URL을 확인하기 위해 두 개의 객체를 선언하였다.

* 생성된 url에 대한 문자열을 매개변수로 받아 단어를 추출하는 함수를 작성하였다. 
* 문자를 추출하는 방식은 < script >, < style >, < url > 등이 쓰인 구간을 삭제하고 HTML태그를 삭제하였다.
* 그 후 &gt;, &lt; 와 같은 것들을 삭제하고 공백을 통해 임시로 단어를 나누었다.
* 임시로 단어를 나눈 것에서 특수문자를 제거해 공백으로 남기고 다시 이 공백을 문자열로 나누고 숫자만으로 구성된 항목을 삭제해 최종적으로 단어 리스트를 반환하였다.


* < script >, < style > 는 단어를 추출하는 함수 외에 또 다른 함수인 delete를 구현하였다.
* 문자열과 시작 문자열, 끝 문자열을 매개변수로 하고 전체 문자열의 처음부터 시작문자열까지, 끝 문자열 이후부터 전체 문자열의 끝까지 반환하는 함수이고 간단히 말하자면 전체 문자열에서 시작 문자열부터 끝 문자열부터 삭제하는 함수이다.
* find()를 통해 시작 문자열과 끝 문자열의 위치를 찾고 발견되었을 경우, 슬라이싱을 이용하여 구간을 삭제한 문자열을 추출한다.
* string = string[:start] + string[end + len(second):] 로 구간을 설정하였는데 find()는 문자열이 시작되는 위치를 찾기 때문에 문자열의 길이만큼을 더해준 곳에서 시작하게 하여 끝 문자열이 포함되지 않도록 하였다.
* 단어를 추출하는 함수인 find_word에는 태그를 제거한 리스트, 공백을 기준으로 나눈 리스트, 특수문자, 숫자 등을 제거한 최종적인 단어를 추출하는 리스트, 단어 저장을 위한 임시 리스트 총 4가지의 리스트를 선언하였다.
* 그리고 위에서 구현한 delete함수의 매개변수로 < script >, < /script> 등을 넣어 해당 구간에 쓰인 코드까지 모두 제거하였다.
* 주석처리된 구간도 해당 문법을 매개변수로 전달하여 제거하였다.


* 그리고 HTML태그를 제거하는데 find함수를 이용하였고 위와 다른 점은 <  > 의 위치를 찾는다는 것이다.
* 태그 내부에 다른 태그가 사용되었을 경우를 확인하였어서 내부 태그도 제거하는 코드를 작성했는데 < < > > 처럼 생긴 경우라면 start는 제일 앞의 < 를 인식하고 end는 세 번째의 > 를 인식하게 된다.
* 그러므로 start와 end 사이의 문자열을 가져와 < > 사이에 < 가 있다면 발견된 <부터 end까지 구간을 삭제하는 작업을 진행하였고 continue를 작성하여 아래 코드가 실행되지 않도록 작성하였다.
* 만약 발견되지 않는다면 슬리이상을 통해 시작 위치전까지의 문자열을 리스트에 추가헉ㅎ 끝나는 위치 +1부터 끝까지 문자열을 추출하였다.
* 위치를 찾고 둘 다 발견되지 않는다면 반복문을 중단하였고 중단되지 않았다면 문자열의 처음부터 시작 위치전까지의 문자열을 리스트에 추가하였고 그 후 슬라이싱을 통해 끝나는 위치 +1 부터 끝까지 문자열을 추출하였다.


* 태그를 제거한 리스트에서 &gt, &lt등의 삭제를 진행하였다.
* 위와 비슷하게 find함수를 사용하였고 & 와 ;의 위치를 찾아 그 구간을 삭제하도록 진행하였다.
* 문자열이 발견되었을 경우에 ;가 단독으로 사용된 경우도 있기 때문에 & 가 ; 보다 뒤에서 발견된다면 다음 항목으로 넘어가도록 하였다.
* 만약 정상적으로 &가 앞에서 ;가 뒤에서 발견이 된다면 list[i] = list[i][:start] + list[i][end + 1:]로 구간 삭제를 진행하였다.
* 문자열의 발견되지 않았을 경우 인덱스를 증가시켜 다음 항목에서 확인할 수 있도록 while문을 지정하였고, i가 리스트의 마지막까지 확인했다면 while문을 중단하도록 하였다.


* 특수문자를 제거하기 전 공백을 통해 임시로 단어를 나누었는데 이는 for문을 이중으로 사용하였다.
* 첫 번째 for문은 list의 크기만큼 반복하고 list의 항목 내에서 공백으로 문자를 잘라 임시 리스트 temp에 저장하였다.
* 두 번째 for문은 이 temp의 크기 만큼 반복되도록 하였고 word라는 리스트에 temp가 모두 들어가도록 append()를 이용해 작성하였다.


* 특수문자 제거는 while문과 함께 isalnum()을 활용하였는데 한글로 구성된 문자열에 isalnum()을 적용하면 True가 나온다는 것을 테스트하였고 문자열이 항목으로 있는 단일 리스트의 경우 list[i][k]를 했을 때 i는 리스트의 항목, k는 문자열이 위치로 인식하는 것 역시 테스트를 통해 확인 후 진행하였다.
* for 문과 그 내부에 while문으로 진행하였는데 for 문은 word리스트의 크기만큼 i가 반복되도록 하였고 k = 0을 선언하여 문자열의 위치를 가리킬 수 있는 변수로 사용하였다.
* while문에서는  if not word[i][k].isalnum(), 즉 리스트의 i번째 항목의 k번째 문자열이 문자나 숫자가 아닌 경우에 if word[i][k] != '' and word[i][k] != ' ', 비어있거나 공백이 아니라면  word[i] = word[i].replace(word[i][k], ' '), 해당 위치의 문자를 공백으로 변경하는 방식으로 진행하였고 만약 위의 두 번째 조건문이 실행되지 않는다면 k 를 증가시켜 문자열의 다음 위치를 확인하였고 첫 번째 조건문이 실행되지 않는 경우 역시 k를 증가시켜 문자열의 다음 위치를 확인할 수 있도록 하였다.
* while문의 중단은 k가 word리스트의 i번째 항목의 길이와 같다면 하나의 항목이 끝난 것이기 때문에 중단시켰다.
* while문 중단 후에는 for 문에서 i가 증가하고 다시 k = 0 으로 초기화되어 다음 항목에 대한 검사를 진행한다.


* 특수문자를 공백으로 바꾸었기 때문에 그 공백에 대해서 다시 문자열을 나누어 result리스트에 추가하였다.
* 나누는 방식은 위에서 한 것 처럼 리스트의 항목을 공백으로 나누어 진행하였다.


* 그리고 마지막으로 반복문과 isdigit()을 확용하여숫자로만 이루어진 항목을 제거하였다.
* while문을 이용하였고 isdigit()이 True가 출력되면 remove()를 통해 삭제되도록 하였고 False가 출력된다면 인덱스 값을 증가시켰다.
* 반복문의 중단은 인덱스가 len(result)와 같다면 반복문을 중단하고 result를 반환하도록 하였다.


* 결과 확인을 위해 함수의 매개변수로 처음에 선언된 객체를 매개변수로 해주었고 함수에서 리턴되는 값을 각각 first, second에 저장하였다.
* 그 후 객체들을 출력하여 단어들을 확인하였고 len()을 이용하여 단어의 개수를 확인하였다.
