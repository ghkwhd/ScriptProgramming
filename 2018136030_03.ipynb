### 4. 다음 문자열을 ':'을 기준으로 분리하여 리스트로 만들고 각 문자열의 좌우 공백을 제거하시오 (즉, 문자열 S에서 l을 만들어라)
    s = '  first star   :   second star   :    third star  '   
    l = ['first star', 'second star', 'third star']   
* 1) for ~ in 문을 이용한 코드를 작성하시오.
* 2) 리스트 내포(list comprehension)를 이용한 코드를 작성하시오.

```python
s = '  first star   :   second star   :    third star  '  
l = s.split(':')

for i in range(len(l)) :
    l[i] = l[i].strip()

print(l)
```

1)
* 문자열을 ' : '를 기준으로 자르기 위해 split(':') 을 이용하였다.
* : 를 기준으로 잘린 리스트의 크기만큼 for문으로 반복시키며 strip()을 통해 각 항목의 좌, 우 공백을 제거하여 다시 그 항목에 대입하였다.

```python
s = '  first star   :   second star   :    third star  '  
l = s.split(':')

l = [i.strip() for i in l]

print(l)
```

2)
* 문자열을 ' : '를 기준으로 자르기 위해 split(':') 을 이용하였다.
* 1) 과는 다르게 반복문은 사용하지만 리스트 내포를 이용한 코드를 작성하는데 이번에는 크기만큼 반복을 시키는 것이 아닌 리스트를 반복시킨다.
* i를 리스트를 반복시키면서 i.strip() 즉, 좌우 공백을 제거한 문자열로 만들어 리스트 l의 항목으로 대입한다.

### 5.  다음과 같이 0보다 큰 정수 리스트 변수 list를 인자로 받는 함수 addall(list)와 addallodd(list)를 for ~ in 리터널과 리스트 내포 방식으로 각각 제시하시오.
* 1) addall(list) 함수 (리스트 내의 모든 정수를 더하는 함수로서 해답으로는 for ~ in 리터럴과 리스트 내포 방식으로 각각 제시하시오.)
     * 
                >>> addall([1])   
                1
     * 
                >>> addall([1, 2, 3, 4, 5, 6, 7, 8, 9])
                45
* 2) addallodd(list) 함수 (리스트내의 모든 홀수를 더하는 함수로서 해답으로는 for ~ in 리터럴과 리스트 내포 방식으로 각각 제시하시오.)   
     * 
                addallodd([1])
                1
     * 
                >>> addallodd([1, 2, 3, 4, 5, 6, 7, 8, 9])
                25
* [참고 1]: 리스트 내포 방식으로 addall() 및 addallodd()함수를 만들 때엔 리스트 내포에 의해 생성된 리스트 결과에 대해 내장 함수를 사용하는 방식을 고려해보시오 (주의: 리스트 내포 방식의 addall() 및 addallodd() 함수의 몸체는 단 1줄로 작성되어야 한다.)
* [참고 2]: sum() 등의 내장함수 사용 가능함

```python
def addall(list) :
    result = []
    for i in list :
        result.append(i)
        
    return sum(result)

print(addall([1, 2, 3, 4]))
```

1-1)
* 리스트 내포 방식을 사용하지 않은 방식이다.
* 함수 내부에 새로운 리스트 result를 선언한다.
* 전달받은 iist를 반복하며 새로운 result 리스트에 항목을 하나씩 추가한다.
* sum()을 이용하여 리스트를 합친 값을 반환하도록 한다.

```python
def addall(list) :
    return sum([i for i in list])

print(addall([1, 2, 3, 4, 5, 6, 7, 8, 9]))
```

1-2)
* 리스트 내포 방식을 사용한 방식이다.
* 함수의 내부에는 이전과 같이 sum()을 이용하여 계산된 값을 반환하는데 새로운 리스트를 생성하는 것이 아니라 sum()내부에서 리스트 내포를 활용한다. 
* list를 반복하며 list의 항목을 새로운 리스트에 대입하여 새로운 리스트를 만드는 코드를 sum() 내부에 선언하여 만들어진 리스트의 합쳐진 값을 반환한다.

```python
def addallodd(list) :
    result = []
    for i in list :
        if i % 2 :
            result.append(i)
        
    return sum(result)

addallodd([1, 2, 3, 4, 5, 6, 7, 8, 9])
```

2-1)
* 리스트 내포 방식을 사용하지 않은 방식이다.
* 함수 내부에 새로운 리스트 result를 선언한다.
* 전달받은 iist를 반복하며 새로운 result 리스트에 항목을 하나씩 추가한다.
* 추가하기 전 홀수만을 추가해야하기 때문에 i % 2를 조건문에 추가하여 2로 나누어지는 경우, 0을 반환하게 되어 실행되지 않도록 한다.
* 만약 나머지가 1인 경우 아래 append()가 실행되어 result리스트에 추가된다.
* sum()을 이용하여 리스트를 합친 값을 반환하도록 한다.

```python
def addallodd(list) :
    return sum([i for i in list if i % 2])

print(addall([1, 2, 3, 4, 5, 6, 7]))
```

2-2)
* 리스트 내포 방식을 사용한 방식이다.
* 함수의 내부에는 이전과 같이 sum()을 이용하여 계산된 값을 반환하는데 새로운 리스트를 생성하는 것이 아니라 sum()내부에서 리스트 내포를 활용한다. 
* list를 반복하며 list의 항목을 새로운 리스트에 대입하는 과정인데 for문 뒤에 조건문 if i % 2를 작성하여 2로 나눈 나머지가 0인 경우(짝수인 경우)에는 항목이 추가되지 않도록 리스트 내포 코드를 작성한다.
* 위 코드를 sum() 내부에서 작성하여 만들어진 리스트의 합을 반환하도록 하였따.

    d['low'] = [10, 20, 30]

* 1) 위 코드의 수행 결과 d와 e는 항상 같은 결과를 출력한다. 그 이유는 무엇인가?

        {'high': [4, 500, 6], 'low': [1, 2, 3]}

```python
L1 = [1, 2, 3]  
L2 = [4, 5, 6]   
d = {'low':L1, 'high':L2}   
e = d   
f = d.copy()   
print(d)   
print(e)   
print(f)   
print()   
d['low'] = [10, 20, 30]  
d['high'][1] = 500   
print(d)   
print(e)   
print(f)   
```

1) 
* e = d를 해주었기 때문에 d를 복사한 e가 생성되는 것이 아니라 e는 d의 별칭이 된다. 그러므로 d와 e의 출력은 항상 똑같이 나오게 된다.

2)
* d.copy()는 얕은 복사이기 때문에 f는 d를 얕은 복사하였다.
* 얕은 복사는 내부의 컬렉션에 대해서는 같은 것을 가리키기 때문에 d의 low가 다른 리스트를 값을 갖는다고해도 d만 변할 뿐 f는 원래의 리스트를 그대로 값으로 갖게 된다.

```python
def isStrange(num):
    '''
    함수 포맷 변경하지 마세요.
    재귀적인 함수는 함수 안에서 자기 자신 함수를 호출하는 것을 말합니다.
    '''
    return True # or False or isStrange(매개변수 숫자)
```

```python
def length(X):
    return len(repr(abs(X)))

def isStrange(num):
    num = int(num)
    if length(num) == 1:
        if num >= 0:
            return True
        else:
            return False
    elif length(num) != 1:
        if num % length(num) == 0:
            return isStrange(num / length(num))
        else:
            return False  # False 반환환


print('0 is Strange? = ', isStrange(0))
print('56 is Strange? = ', isStrange(56))
print('100 is Strange? = ', isStrange(100))
print('7882602506492313600 is Strange? = ', isStrange(7882602506492313600))
print('6012302549522098651 is Strange? = ', isStrange(6012302549522098651))
```

* [예시]

```python
def findDivisor(n, num):
    divisors = []
    result = []

    for i in range(1, int(n ** 0.5) + 1):
        if n % i == 0:
            divisors.append(i)
            if i != (n // i):
                divisors.append(n // i)
    divisors.sort()

    for i in divisors :
        if i >= num:
            result.append(i)

    return result

def solve(a):
    bus = []

    divisor = findDivisor(sum(a), max(a))

    for i in divisor:
        temp = 0
        for j in a:
            temp += j
            if temp == i:
                temp = 0
            elif temp > i:
                break

        if temp == 0 and j == a[len(a)-1]:
            bus.append(i)

    return bus

a = [9999] * 10 ** 5
num = solve(a)
print('가능한 버스의 좌석 수 = ', num)
```

```python
import string
import requests

# 전달받은 문자열 사이의 구간 삭제
def delete(string, first, second):
    while (True):
        start = string.find(first)  # first에 들어온 단어의 위치 확인
        end = string.find(second)  # end로 들어온 단어의 위치 확인
        if start != -1 and end != -1:
            string = string[:start] + string[end + len(second):]  # end는 시작되는 위치가 나오기 때문에 end는 글자수 만큼 더해준다
        elif start == -1 and end == -1:
            break
    return string

def find_word(str):
    list = []  # 태그를 제거한 리스트
    word = []  # 공백을 기준으로 나눈 리스트
    result = []  # 특수문자, 숫자, 기타 문법을 제거한 최종 리스트
    dic = { }
    temp = []  # 단어 저장을 위한 임시 리스트

    # 태그 사이에 문법이 들어가는 경우의 삭제
    # 간혹 대문자로 쓰인 페이지가 있어서 대문자로 이루어진 태그도 구간 삭제
    str = delete(str, '<style', '</style>')  # <style ~ </style> 구간 삭제
    str = delete(str, '<STYLE', '</STYLE>')  # <STYLE ~ </STYLE> 구간 삭제
    str = delete(str, '<script', '</script>')  # <script ~ </script> 구간 삭제
    str = delete(str, '<SCRIPT', '</SCRIPT>')  # <SCRIPT ~ </SCRIPT> 구간 삭제
    str = delete(str, '<!--', '-->')  # 주석처리 구간 삭제

    # HTML 태그 제거
    while (True):
        start = str.find('<')  # <가 있는 위치 확인
        end = str.find('>')  # >가 있는 위치 확인
        if start == -1 and end == -1:  # 만약 < 와 > 가 없다면
            break  # 모든 문자열을 확인한 것이니 반복문 중단
        # <> 내부에 <>가 있는지 확인
        inner = str[start + 1: end+1]     # <>사이를 자른 문자열 생성
        in_start = inner.find('<')  # <가 있는 위치 확인
        if in_start != -1 :      # <> 사이에 <가 있다면 < <> > 형태이므로
            str = str[:in_start] + str[end + 1:]    # 내부의 <> 삭제
            continue
        list.append(str[:start])  # 반복문이 중단되지 않았다면 처음부터 start 전까지의 문자열을 list에 저장
        str = str[end + 1:]  # > 이후의 문자열만 남긴다

    # &gt, &lt 등 삭제
    i = 0
    while (True):
        start = list[i].find('&')  # 시작이 &
        end = list[i].find(';')  # 끝이 ;
        if start != -1 and end != -1:  # 둘 다 발견되었을 경우
            if start < end :    # &가 앞에서, ;가 뒤에서 발견된 경우
                list[i] = list[i][:start] + list[i][end + 1:]  # &와 ; 사이의 구간을 삭제
            else :      # ; 하나만 쓰여서 &가 뒤에서 발견되는 경우 (나중에 특수문자에서 제거된다)
                i += 1  # 다음 항목으로 넘어간다
        else:  # 발견되지 않았을 경우
            i += 1  # 다음 단어 확인

        if i == len(list):  # 마지막 단어까지 확인하면 중단
            break

    # 공백을 통해 임시로 단어 나누기
    for i in range(len(list)):
        temp = list[i].split()  # 리스트의 항목을 반복시켜 공백으로 나누고 임시 리스트에 저장
        for j in range(len(temp)):  # 임시 리스트의 항목 수 만큼 반복
            word.append(temp[j])  # 임시 리스트의 내용들을 word 리스트에 저장

    # string.punctuation을 이용하여 구두문자 제거
    for i in word :
        i = i.replace(string.punctuation, ' ')

    # string.punctuation으로 제거되지 않은 특수문자 제거
    for i in range(len(word)):
        k = 0
        while (True):
            if not word[i][k].isalnum():  # 문자(영어, 한글)나 숫자로만 이루어진 경우가 아닐 때
                if word[i][k] != '' and word[i][k] != ' ':  # 아무것도 없는 항목이거나 공백만 들어있는 것이 아닐 때
                    word[i] = word[i].replace(word[i][k], ' ')  # 해당 위치를 공백으로 변경
                else:  # 아무것도 들어있지 않거나 공백이라면
                    k += 1  # 단어의 다음 위치 확인
            else:  # 문자와 숫자로만 이루어진 경우
                k += 1  # 문자의 다음 위치 확인

            if k == len(word[i]):  # 문자열의 마지막 부분까지 확인했다면
                break  # 반복 중단

    # 구두문자 제거 이후 공백 항목 제거
    for i in range(len(word)):
        temp = word[i].split()  # word 항목을 반복시켜 공백으로 나누고 임시 리스트에 저장
        for j in range(len(temp)):  # 임시 리스트의 항목 수 만큼 반복
            result.append(temp[j])  # 임시 리스트의 내용들을 result 리스트에 저장

    # 숫자만으로 이루어진 항목 제거
    i = 0
    while (True):
        if result[i].isdigit():  # 항목이 숫자로만 구성되어 있다면 True를 반환해서 조건문 실행
            result.remove(result[i])  # 숫자만으로 구성된 항목은 삭제
        else:
            i += 1

        if i == len(result):
            break

    # 각 단어의 출현 빈도를 사전형태로 저장
    for i in result :
        dic[i] = dic.get(i, 0) + 1

    return dic

urlList = ['http://cse.koreatech.ac.kr', 'https://www.koreatech.ac.kr', 'http://www.naver.com', 'http://www.daum.net','http://www.nytimes.com']

for i in urlList :
    req = requests.get(i)
    source = req.text
    words = find_word(source)
    print(i)
    print(words)
    print("\n\n")
```

* HTML태그와 &gt와 같은 태그까지 모두 제거하고 공백으로 단어를 나눈 리스트에서 구두문자 제거를 하였습니다.
* import string을 통해 string.punctuation를 사용할 수 있었고, replace()를 이용하여 string.punctuation에 포함된 구두문자들을 공백으로 대체하였습니다.
* print(string.punctuation)로 확인을 했을 때, 모든 구두문자가 제거되지 않아 이전에 특수문자를 제거하는 코드를 이용하여 제거되지 않은 특수문자를 제거하였습니다.
* 구두문자/특수문자 제거 이후 공백으로 판명되는 항목의 제거 역시 지난 번 과제에서 동일하게 수행했던 내용입니다. word리스트를 반복하며 리스트를 공백으로 나누어 임시 리스트에 넣고 그 임시 리스트의 길이만큼 반복하며 최종 결과 리스트에 추가하였습니다.
* 문자의 출현 빈도를 나타내기 위해 dic[i] = dic.get(i, 0) + 1를 사용하였는데, 0을 기본 값으로 하여 전달받은 문자열이 신규 단어라면 이 기본값 0을 반환하고 1을 추가해 딕셔너리에 해당 단어를 키로 하여 값 1을 추가하고 이미 존재하는 단어이면 저장되어 있는 출현 빈도수가 get으로 반환되어 1을 더해 빈도수가 증가하도록 하였습니다.
* 5개의 링크에 대한 결과를 확인하기 위해 5개의 링크를 리스트에 추가하였고, 이 리스트를 반복시켜 함수에 전달하는 방식으로 코드를 작성하였습니다.
