## 2. 다형성에 대해 설명하고 다형성을 보여주는 자신만의 파이썬 코드 예제를 제시하시오.

* 상속 관계 내의 다른 클래스가 같은 이름의 메소드 호출에 대해 서로 다르게 동작하도록 하는 기능을 다형성이라고 한다.
<br>
<br>
* 아래의 예시를 보면 클래스 A에는 cal이라는 메소드가 정의되어 있다.
* Add 클래스와 Sub 클래스는 모두 동일하게 A 클래스를 상속받고, cal메소드를 서로 다르게 재작성하였다.
* A클래스, Add클래스, Sub클래스의 객체를 생성하고 모든 객체들에 대해 cal 메소드를 호출하면 모든 객체가 다르게 동작하는 것을 볼 수 있다.

```python
class A:
    def __init__(self, x, y):
        self.x = x
        self.y = y
        
    def cal(self):
        print("계산 메소드")
        
class Add(A):
    def __init__(self, x, y):
        super(Add, self).__init__(x,y)
        
    def cal(self):
        print("더하기 메소드")
        print(self.x, '+', self.y, '=',self.x + self.y)
        
class Sub(A):
    def __init__(self, x, y):
        super(Sub, self).__init__(x,y)
        
    def cal(self):
        print("빼기 메소드")
        print(self.x, '-', self.y, '=', self.x - self.y)
        
a = A(7,2)
b = Add(7,2)
c = Sub(7,2)

a.cal()
print()
b.cal()
print()
c.cal()
```

## 3. 다음 각 요구사항 모두를 만족시키는 Counter 클래스를 코딩하시오 (정답을 각 요구사항별로 입력할 필요 없이 3번 문제에 대해 1개의 클래스 정의 코드를 제시하면 된다.)

* 요구사항 5. 다음과 같은 두 개의 산술 연산 (+, -)이 수행될 수 있도록 Counter 클래스 내에 관련 메소드를 추가하시오.

>\>>> c = c + 5   
>\>>> d = d - 5   
>\>>> print(c)   
>[Count (step: 1)] 17   
>\>>> print(d)   
>[Count (step: 2)] 9   

* 요구사항 6. 다음과 같은 관계연산 (>, <, ==)이 수행될 수 있도록 Counter 클래스 내에 관련 메소드를 추가하시오.

>\>>> print(c > 10)   
True   
\>>> print(d > 10)   
False   
\>>> print(c < 10)  
False  
\>>> print(d < 10)  
True  
\>>> print(c == 17)  
True  
\>>> print(d != 9)  
False  

```python
class Counter:
    def __init__(self, count, step = 1): # 요구사항 1번
        self.count = count
        self.step = step
        
    def __str__(self): # 요구사항 2번
        return '[{} (step: {})] {}'.format(self.__class__.__name__, self.step, self.count)
    
    def incr(self): # 요구사항 3번
        self.count += self.step
        
    def __call__(self): # 요구사항 4번
        self.incr()
        
    def __add__(self, num): # 요구사항 5번
        return Counter(self.count + num, self.step)
    
    def __sub__(self, num): # 요구사항 5번
        return Counter(self.count - num, self.step)
    
    def __gt__(self, num): # 요구사항 6번
        return self.count > num
    
    def __lt__(self, num): # 요구사항 6번
        return self.count < num
    
    def __eq__(self, num): # 요구사항 6번
        return self.count == num
    
    def __ne__(self, num): # 요구사항 6번
        return self.count != num
        
c = Counter(10)
d = Counter(10,2)

print(c)
print(d)

c.incr()
d.incr()

print(c)
print(d)

c()
d()

print(c)
print(d)

c = c + 5
d = d - 5

print(c)
print(d)

print(c>10)
print(d > 10)
print(c < 10)
print(d < 10)
print(c == 17)
print(d != 9)
```

* 요구사항 1번

    * 인자가 하나가 들어가면 count 값에 들어가도록 step = 1로 작성하여 step의 디폴트 값을 설정
    * 전달된 값이 들어갈 수 있도록 self를 활용하여 값을 초기화한다.

* 요구사항 2번

    * 주어진 양식에 맞게 작성하고 format을 이용하여 값을 집어넣었으며 그 문자열을 return 하도록 하였다.

* 요구사항 3번
    * incr()은 count값을 step만큼 증가시키기 때문에 객체의 count값인 self.count의 현재 값에 객체의 step값인 self.step을 더하도록 코드를 작성한다.

* 요구사항 5번

    * \__add__가 호출되었을 때 호출 결과로 신규 Count값이 되도록 하기 위해 self.count에 전달받은 num값을 더해준 값과 현재의 step값으로 객체를 생성하고 이를 리턴한다.

## 4. 다음은 내장 자료형 list를 서브클래싱하여 만든 MySet 클래스 정의 내용이다. 다음 클래스 정의에서 __init__(), __str()__(), elimicate_duplicate()의 세 개의 메소드 코드 내용을 자신이 다른 사람에게 가르친다고 생각하며 설명해보시오.

* MySet은 집합(Set) 자료형을 정의하려는 의도하에 만들어진 클래스이다.

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)

    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print(s)
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print(t)
```

* staticmethod는 인스턴스를 만들지 않아도 사용할 수 있는 메소드이다.
* eliminate_duplicatd()를 보면 리스트 내의 중복된 항목을 제거하는 코드인데 집합 자료형은 중복을 허용하지 않기 때문에 위와 같은 메소드가 필요한데 이 메소드는 객체 생성 전에 전달받은 리스트에서 중복 항목을 제거하기 위해  staticmethod로 선언되었고, 클래스이름을 통해 접근 가능하다.
* \__init__()은 생성자 함수인데 객체가 생성될 때 자동으로 호출되는 메소드이다. 집합 자료형 객체를 생성한다면 이 메소드가 호출되고 생성자 내의 코드가 자동으로 생성되어 위 코드에서 결과적으로 중복 항목이 없는 객체가 생성된다.
* \__str__()은 객체를 print()할 시에 호출되는 메소드인데 사용자가 원하는 형태로 객체 데이터를 담은 문자열을 반환한다. 위의 코드에서는 원하는 문자열과 객체의 정보들을 문자열로 바꾸어 result에 선언하고 이를 반환하였다.

## 5. 4번 문제에 정의된 MySet 클래스에 메소드를 추가하여 다음 각 요구사항 모두를 만족시키는 코딩을 제시하시오.

* 정답을 각 요구사항별로 입력할 필요 없이 요구사항 3개 전체에 대해 1개의 MySet 클래스 정의 코드를 제시하고 주석등으로 해당 코드를 설명한다.
* 요구사항 1. | 연산으로 두 집합의 합집합을 반환한다.

>\>>> u = s | t   
>\>>> print(u)   
>MySet: {1, 2, 3, 4, 5, 6, 7, 8, 9}   

* 요구사항 2. & 연산으로 두 집합의 교집합을 반환한다.

>\>>> u = s & t   
>\>>> print(u)   
>MySet: {2, 3}   

* 요구사항 3. - 연산으로 두 집합의 차집합을 반환한다.

>\>>> s = MySet([1, 2, 3])   
>\>>> t = MySet([3, 4, 5])   
>\>>> u = s - t   
>\>>> print(u)   
>MySet: {1, 2}   

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e) # 전달받은 항목을 반복하며 객체에 추가
        MySet.eliminate_duplicate(self) # 중복항목을 없애기 위해 staticmethod호출
    
    def __str__(self): # 객체를 출력할 때 출력 형태 지정
        result = "MySet: {"
        for e in self: # 객체가 가진 항목을 문자열로 표현
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result # 문자열 반환

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s: # 중복항목이 아닌 경우 추가한다
                s.append(e)
        l[:] = [] # 리스트 초기화
        for e in s: # 중복 제거된 항목들을 추가
            l.append(e)
    
    def __or__(self,other): # 요구사항 1번
        new = self + other # new의 타입은 리스트
        return MySet(new) # 합해진 리스트로 객체를 생성하여 반환, 객체를 생성하면 자동으로 중복 제거 
    
    def __and__(self,other): # 요구사항 2번
        new = [x for x in self if x in other] # self를 반복하며 other에 있는 항목을 리스트에 추가
        return MySet(new) # new 리스트로 객체를 만들어 반환
    
    def __sub__(self, other): # 요구사항 3번
        new = [x for x in self if x not in other] # 차집합이기 때문에 self를 반복하며 other에 있지 않은 항목을 리스트에 추가
        return MySet(new) # new 리스트로 객체를 만들어 반환

    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print(s)
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print(t)
    
    u = s | t
    print('합집합 -> ', u)
    
    u = s & t
    print('교집합 -> ', u)
    
    s = MySet([1, 2, 3])
    t = MySet([3, 4, 5])
    u = s - t
    print('차집합 -> ', u)
```

* 요구사항 1번
    * 객체에 대한 |를 활용하기 위해 \__or__ 메소드를 정의하였다.
    * 생성자에서 eliminate_duplicate를 호출하기 때문에 어떤 리스트를 전달해도 중복이 제거된 MySet객체가 만들어지게 된다. 이 점을 이용하여 합집합을 쉽게 구현할 수 있다.
    * 인자로 전달받은 other와 self를 합하여 new 라는 리스트를 생성한다. (new의 type을 확인해보았는데 리스트였다)
    * 이 리스트로 MySet객체를 만들면 생성자에서 중복을 제거하기 때문에 만들어진 객체를 바로 반환한다.

* 요구사항 2번
    * 객체에 대한 &를 활용하기 위해 \__and__ 메소드를 정의하였다.
    * u에 교집합의 결과를 참조해야하기 때문에 return 시 객체를 리턴할 수 있도록 하였다.
    * 교집합은 self를 반복하며 반복되는 것이 other에 있다면 리스트에 추가하였고 이 리스트로 MySet 객체를 만들어 반환하였다.

* 요구사항 3번
    * 객체에 대한 -를 활용하기 위해 \__sub__ 메소드를 정의하였다.
    * u에 차집합의 결과를 참조해야하기 때문에 return 시 객체를 리턴할 수 있도록 하였다.
    * 차집합은 self를 반복하며 반복되는 것이 other에 있지 않다면 리스트에 추가하였고 이 리스트로 MySet 객체를 만들어 반환하였다.

## 6. 5번 문제에서 정의한 MySet 클래스에 대해 다음 예제를 수행하면 오류없이 올바르게 동작하는 것을 확인할 수 있다. 다음 예제 내에 있는 len(), bool() 내장함수와 in 키워드 사용 예제가 별다른 메소드 정의를 하지 않았는 데도 올바르게 수행되는 이유를 설명하시오.

>\>>>s = MySet([1, 2, 3, 4, 5, 6])    
\>>>print(len(s))   
6<br/>
\>>>print(bool(s))    
True    
\>>>print(2 in s)    
True   

* 매직메소드는 다른 메소드와 같은 방식으로 호출되지만, 특정 조건에 따라 자동으로 호출되기도 한다.
* bool()은 \__bool__(self)가 호출되는데 이 매직메소드는 정의하지 않으면 True가 기본적으로 호출되기 때문에 정상적으로 작동한다.
* len()은 \__len__(self)가 호출되는데 이 매직메소드는 컬렉션 클래스 매직메소드이다. 위의 MySet 클래스를 보면 self 자체가 리스트이기 때문에 따로 정의하지 않아도 수행이 가능한 것이라고 생각된다.

* 파라미터로 받은 단어와 유사도가 동일한 URL이 여러 개이면 해당 URL을 모두 출력하시오.

```python
import string
import requests

class SearchEngine():
    req = requests.get('https://raw.githubusercontent.com/stopwords-iso/stopwords-ko/master/stopwords-ko.txt')
    source = req.text
    stop_word = source.split()

    with open('stop_words_english.txt', 'r', encoding='UTF-8') as file:
        file_list = file.readlines()

    for i in range(len(file_list)):
        file_list[i] = file_list[i].rstrip('\n')

    stop_word += file_list

    def __init__(self, *args):
        self.url = []
        self.dic = {} # 단어들의 사전
        self.dic_list = []  # url별 단어 출현 빈도
        self.sorted_list = []
        self.frequency = [] # 최대 빈도 수 단어
        for i in range(len(args)):
            self.url.append(args[i])

    def addUrl(self, item):
        self.url.append(item)

    def removeUrl(self, item):
        if item in self.url:
            self.url.remove(item)

    def listUrls(self):
        print('<URL List>')
        for i in self.url:
            print(i)
        print()

    def delete(self, string, first, second):
        while (True):
            start = string.find(first)
            end = string.find(second)
            if start != -1 and end != -1:
                string = string[:start] + string[end + len(second):]
            elif start == -1 and end == -1:
                break
        return string


    def getWordsFrequency(self):
        for i in self.url:
            req = requests.get(i)
            source = req.text

            list = []  # 태그를 제거한 리스트
            word = []  # 공백을 기준으로 나눈 리스트
            result = []  # 특수문자, 숫자, 기타 문법을 제거한 최종 리스트
            temp = []  # 단어 저장을 위한 임시 리스트
            dic = {}

            source = self.delete(source, '<style', '</style>')
            source = self.delete(source, '<STYLE', '</STYLE>')
            source = self.delete(source, '<script', '</script>')
            source = self.delete(source, '<SCRIPT', '</SCRIPT>')
            source = self.delete(source, '<!--', '-->')

            # HTML 태그 제거
            while (True):
                start = source.find('<')
                end = source.find('>')
                if start == -1 and end == -1:
                    break
                # <> 내부에 <>가 있는지 확인
                inner = source[start + 1: end + 1]
                in_start = inner.find('<')
                if in_start != -1:
                    source = source[:in_start] + source[end + 1:]
                    continue
                list.append(source[:start])
                source = source[end + 1:]

            # &gt, &lt 등 삭제
            i = 0
            while (True):
                start = list[i].find('&')
                end = list[i].find(';')
                if start != -1 and end != -1:
                    if start < end:
                        list[i] = list[i][:start] + list[i][end + 1:]
                    else:
                        i += 1
                else:
                    i += 1

                if i == len(list):
                    break

            # 공백을 통해 임시로 단어 나누기
            for i in range(len(list)):
                temp = list[i].split()
                for j in range(len(temp)):
                    word.append(temp[j])

            # sourceing.punctuation을 이용하여 구두문자 제거
            for i in word:
                i = i.replace(string.punctuation, ' ')

            # sourceing.punctuation으로 제거되지 않은 특수문자 제거
            for i in range(len(word)):
                k = 0
                while (True):
                    if not word[i][k].isalnum():
                        if word[i][k] != '' and word[i][k] != ' ':
                            word[i] = word[i].replace(word[i][k], ' ')
                        else:
                            k += 1
                    else:
                        k += 1

                    if k == len(word[i]):
                        break

            # 구두문자 제거 이후 공백 항목 제거
            for i in range(len(word)):
                temp = word[i].split()
                for j in range(len(temp)):
                    result.append(temp[j])

            # 숫자만으로 이루어진 항목 제거
            i = 0
            while (True):
                if result[i].isdigit():
                    result.remove(result[i])
                else:
                    i += 1

                if i == len(result):
                    break

            # 각 단어의 출현 빈도를 사전형태로 저장
            for i in result:
                dic[i] = dic.get(i, 0) + 1

            # 불용어 제거
            for i in SearchEngine.stop_word:
                if i in dic:
                    del dic[i]

            # 영어 불용어를 대, 소문자에 관계없이 제거하기 위해 선언
            engKeys = []
            for k in dic.keys():
                str_key = str(k)
                if str_key.encode().isalpha():
                    engKeys.append((k, str_key.lower()))

            for i in SearchEngine.stop_word:
                for j in engKeys:
                    if i == j[1]:
                        del dic[j[0]]

            self.dic_list.append(dic)   # url 별 단어 출현 빈도 사전 리스트

            # 하나의 사전으로 합치기
            for i in dic:
                if i in self.dic:
                    self.dic[i] = self.dic[i] + dic[i]
                else:
                    self.dic[i] = dic[i]

        return self.dic


    def getMaxFreqencyWords(self):
        self.sorted_list = sorted(self.dic.items(), key=lambda x: x[1], reverse=True)

        for i in range(len(self.sorted_list)):
            self.frequency.append(self.sorted_list[i])
            if i == len(self.sorted_list)-1:
                break
            if self.sorted_list[i][1] != self.sorted_list[i+1][1]:
                break

        return self.frequency


    def similarity(self, search, dic, num):
        similar = 0.0
        if search in dic:
            value = dic[search]
            similar += value / num  # num : 단어들의 빈도 수의 합

        for j in dic.keys():
            str_keys = str(j)
            if search in str_keys:  # 단어 포함 여부를 판단으로 한 유사도 판단
                value = dic[j]
                similar += (value / num) * 0.5  # 포함 여부는 검색의 정확성이 일치여부보다 낮기 때문에 0.5를 곱해준다

        return similar

    def searchUrlByWord(self, search):
        similarityDict = {}
        sumWords = []
        for i in self.dic_list:
            sum = 0
            for j in i.values():
                sum += 1
            sumWords.append(sum)

        for i in range(len(self.url)):
            similar = self.similarity(search, self.dic_list[i], sumWords[i])  # 유사도 판단
            similarityDict[self.url[i]] = str(similar)  # 사전에 추가

        sorted_similar = sorted(similarityDict.items(), key=lambda x: x[1], reverse=True)  # 사전을 값의 내림차순 형태로 정렬

        print('< {} >에 대한 유사도가 높은 url'.format(search))
        for i in range(len(sorted_similar)):
            print(sorted_similar[i][0])
            if i == len(sorted_similar)-1:
                break
            if sorted_similar[i][1] != sorted_similar[i+1][1]:
                break
            j += 1

w1 = SearchEngine('http://www.cnn.com', 'http://www.times.com', 'https://www.amazon.com')
w2 = SearchEngine('http://www.cnn.com', 'http://www.times.com')
w3 = SearchEngine()

w1.addUrl('https://github.com')
w3.addUrl('http://stackoverflow.com')

w1.removeUrl('http://www.cnn.com')
w2.removeUrl('http://stackoverflow.com')

w1.listUrls()

print('<w1의 단어 출현 빈도 사전>')
print(w1.getWordsFrequency())
print()
print('<w1의 최대 출현 빈도 단어 리스트>')
print(w1.getMaxFreqencyWords())
print()
w1.searchUrlByWord("news")
```

```python
import string
import requests

class SearchEngine():
    req = requests.get('https://raw.githubusercontent.com/stopwords-iso/stopwords-ko/master/stopwords-ko.txt')
    source = req.text
    stop_word = source.split()

    with open('stop_words_english.txt', 'r', encoding='UTF-8') as file:
        file_list = file.readlines()

    for i in range(len(file_list)):
        file_list[i] = file_list[i].rstrip('\n')

    stop_word += file_list

    def __init__(self, *args):
        self.url = []
        self.dic = {} # 단어들의 사전
        self.dic_list = []  # url별 단어 출현 빈도
        self.sorted_list = []
        self.frequency = [] # 최대 빈도 수 단어
        for i in range(len(args)):
            self.url.append(args[i])

    def addUrl(self, item):
        self.url.append(item)

    def removeUrl(self, item):
        if item in self.url:
            self.url.remove(item)

    def listUrls(self):
        print('<URL List>')
        for i in self.url:
            print(i)
        print()

    def delete(self, string, first, second):
        while (True):
            start = string.find(first)
            end = string.find(second)
            if start != -1 and end != -1:
                string = string[:start] + string[end + len(second):]
            elif start == -1 and end == -1:
                break
        return string


    def getWordsFrequency(self):
        for i in self.url:
            req = requests.get(i)
            source = req.text

            list = []  # 태그를 제거한 리스트
            word = []  # 공백을 기준으로 나눈 리스트
            result = []  # 특수문자, 숫자, 기타 문법을 제거한 최종 리스트
            temp = []  # 단어 저장을 위한 임시 리스트
            dic = {}

            source = self.delete(source, '<style', '</style>')
            source = self.delete(source, '<STYLE', '</STYLE>')
            source = self.delete(source, '<script', '</script>')
            source = self.delete(source, '<SCRIPT', '</SCRIPT>')
            source = self.delete(source, '<!--', '-->')

            # HTML 태그 제거
            while (True):
                start = source.find('<')
                end = source.find('>')
                if start == -1 and end == -1:
                    break
                # <> 내부에 <>가 있는지 확인
                inner = source[start + 1: end + 1]
                in_start = inner.find('<')
                if in_start != -1:
                    source = source[:in_start] + source[end + 1:]
                    continue
                list.append(source[:start])
                source = source[end + 1:]

            # &gt, &lt 등 삭제
            i = 0
            while (True):
                start = list[i].find('&')
                end = list[i].find(';')
                if start != -1 and end != -1:
                    if start < end:
                        list[i] = list[i][:start] + list[i][end + 1:]
                    else:
                        i += 1
                else:
                    i += 1

                if i == len(list):
                    break

            # 공백을 통해 임시로 단어 나누기
            for i in range(len(list)):
                temp = list[i].split()
                for j in range(len(temp)):
                    word.append(temp[j])

            # sourceing.punctuation을 이용하여 구두문자 제거
            for i in word:
                i = i.replace(string.punctuation, ' ')

            # sourceing.punctuation으로 제거되지 않은 특수문자 제거
            for i in range(len(word)):
                k = 0
                while (True):
                    if not word[i][k].isalnum():
                        if word[i][k] != '' and word[i][k] != ' ':
                            word[i] = word[i].replace(word[i][k], ' ')
                        else:
                            k += 1
                    else:
                        k += 1

                    if k == len(word[i]):
                        break

            # 구두문자 제거 이후 공백 항목 제거
            for i in range(len(word)):
                temp = word[i].split()
                for j in range(len(temp)):
                    result.append(temp[j])

            # 숫자만으로 이루어진 항목 제거
            i = 0
            while (True):
                if result[i].isdigit():
                    result.remove(result[i])
                else:
                    i += 1

                if i == len(result):
                    break

            # 각 단어의 출현 빈도를 사전형태로 저장
            for i in result:
                dic[i] = dic.get(i, 0) + 1

            # 불용어 제거
            for i in SearchEngine.stop_word:
                if i in dic:
                    del dic[i]

            # 영어 불용어를 대, 소문자에 관계없이 제거하기 위해 선언
            engKeys = []
            for k in dic.keys():
                str_key = str(k)
                if str_key.encode().isalpha():
                    engKeys.append((k, str_key.lower()))

            for i in SearchEngine.stop_word:
                for j in engKeys:
                    if i == j[1]:
                        del dic[j[0]]

            self.dic_list.append(dic)   # url 별 단어 출현 빈도 사전 리스트

            # 하나의 사전으로 합치기
            for i in dic:
                if i in self.dic:
                    self.dic[i] = self.dic[i] + dic[i]
                else:
                    self.dic[i] = dic[i]

        return self.dic


    def getMaxFreqencyWords(self):
        self.sorted_list = sorted(self.dic.items(), key=lambda x: x[1], reverse=True)

        for i in range(len(self.sorted_list)):
            self.frequency.append(self.sorted_list[i])
            if self.sorted_list[i][1] != self.sorted_list[i+1][1]:
                break

        return self.frequency


    def similarity(self, search, dic, num):
        similar = 0.0
        if search in dic:
            value = dic[search]
            similar += value / num  # num : 단어들의 빈도 수의 합

        for j in dic.keys():
            str_keys = str(j)
            if search in str_keys:  # 단어 포함 여부를 판단으로 한 유사도 판단
                value = dic[j]
                similar += (value / num) * 0.5  # 포함 여부는 검색의 정확성이 일치여부보다 낮기 때문에 0.5를 곱해준다

        return similar

    def searchUrlByWord(self, search):
        similarityDict = {}
        sumWords = []
        for i in self.dic_list:
            sum = 0
            for j in i.values():
                sum += 1
            sumWords.append(sum)

        for i in range(len(self.url)):
            similar = self.similarity(search, self.dic_list[i], sumWords[i])  # 유사도 판단
            similarityDict[self.url[i]] = str(similar)  # 사전에 추가

        sorted_similar = sorted(similarityDict.items(), key=lambda x: x[1], reverse=True)  # 사전을 값의 내림차순 형태로 정렬

        print('< {} >에 대한 유사도가 높은 url'.format(search))
        j = 0
        for i in sorted_similar:
            print(i[0])
            if j == len(i)-1:
                break
            if i[j][1] != i[j+1][1]:
                break
            j += 1

#######################################################################################################

# 소문자 true와 false를 사용하기 위해 선언
false = 1
true = 0

class SearchEngineWithOrderedWebWords(SearchEngine):
    def __init__(self, *args):
        super(SearchEngineWithOrderedWebWords, self).__init__(*args)
        super().getWordsFrequency()

    def getWordsFrequency(self, reverse=false):
        self.sorted_list = sorted(self.dic.items(), key=lambda x: x[1], reverse=reverse)
        return self.sorted_list

w4 = SearchEngineWithOrderedWebWords('http://www.times.com', 'https://www.amazon.com', 'https://github.com')
print(w4.getWordsFrequency())
print()
print(w4.getWordsFrequency(reverse=true))
```

* 자식 클래스의 getWordsFrequency()에서는 단어들의 출현 빈도를 가진 사전에 대한 정렬만 진행하여 출력하기 위해 자식 클래스에서 부모클래스의 getWordsFrequency()를 호출해주었다.
* true와 false를 사용하기 위해 false = 1, true = 0 으로 정의하였다.
* 위와 같이 정의한 이유는 getWordsFrequency()의 인자 reverse는 true와 false를 인자로 받게 되는데 원래 파이썬에서는 True, False로 나타내는데 소문자 false, true를 사용하기 위해 선언하였다.
* 또 reverse가 true일 때 오름차순으로 정렬되어야하는데 sorted의 reverse 인자가 False일 때 오름차순 정렬을 하기 때문에 오름차순으로 정렬해야하는 true의 값을 0으로, false의 값을 1로 설정하고 sorted의 reverser 인자에 getWordsFrequency()의 reverse인자를 넣어주었다.

* 객체를 반복했을 때 단어들이 정렬된 리스트를 순차적으로 출력될 수 있도록 \__iter__()와 \__next__()를 정의하였다.
* \__iter__()에는 self.current = 0 이라는 코드를 작성하였는데 이는 어디까지 출력이 되었는지 알 수 있도록 하기 위해서이다.
* \__next__()에는 self.current가 self.sorted_list의 길이보다 작다면 self.current를 1 증가시키고 self.current-1의 인덱스에 해당하는 값을 반환하도록 하였고, 그게 아니라면 StopIteration을 발생시키도록 하였다.
